```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
a=pd.read_csv('climate_change_food_supply_india.csv')
a.head()
plt.figure(figsize=(10,7))
avg_yield= a.groupby("Crop Type")["Crop Yield (ton/ha)"].mean()
plt.bar(x=avg_yield.index, height=avg_yield.values, color="royalblue", alpha=0.5)
plt.xlabel("Crop Types", fontsize=13, weight="bold")
plt.ylabel("Crop Yield", fontsize=13, weight="bold")
plt.xticks(rotation=45, ha="right")
plt.title("Comparison Of Average Crop Yields Across Different Crop Types", fontsize=15, weight="bold")
plt.show()

#SugarCane Performs Best
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.figure(figsize=(10,7))
avg_rainfall= a.groupby("Region")['Rainfall (mm)'].mean()
plt.bar(x=avg_rainfall.index, height=avg_rainfall.values, color="blue", alpha=0.5)
plt.xlabel("Regions", fontweight="bold", size=13)
plt.ylabel("Rainfall", fontweight="bold", size=13)
plt.xticks(rotation=45, ha='right')
plt.title("Rainfall Levels Vary Across The Different Regions.", fontweight="bold", size=15)
plt.show()

#Rajasthan has more rainfall
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.figure(figsize=(10,7))
a=pd.read_csv('climate_change_food_supply_india.csv')
a.head()
plt.scatter(x=a['Avg Temp (°C)'], y=a['Crop Yield (ton/ha)'], alpha=0.4, color="green", marker="s")

plt.annotate("High Temperature results in Less Crop Productivity",
             xy=(26, 7.2),
             xytext=(26.5, 7.2), 
             arrowprops={'arrowstyle':'->', 'color':'green'},
             fontsize=10, color="blue", alpha=0.8,  fontweight="medium")

plt.grid(True, axis="both") 
plt.xlabel("Average Temperature", fontweight="bold", size=13)
plt.ylabel("Crop Yield", fontweight="bold", size=13)
plt.title("Relationship Between Average Temperature and Crop Yield", fontweight="bold", size=15)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_csv("climate_change_food_supply_india.csv")
severity_counts = a.groupby(['Year', 'Severity Level']).size().unstack(fill_value=0)
severity_counts.plot(kind="bar",stacked=True,
                      figsize=(10, 6), color={"Low": "yellow", "Medium": "orange", "High": "red"}, width=0.9)
plt.xlabel("Year", fontweight="bold", size=13)
plt.ylabel("Number of Events", fontweight="bold", size=13)
plt.title("Severity of Climate Events Over the Years", fontweight="bold", size=15)
plt.legend()
plt.xticks(rotation=45)
plt.show()

#The stacked bar chart indicates that the year 2020 faced the highest number of severe climate events
```

```python
import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_csv("climate_change_food_supply_india.csv")
price_data = a[a['Year'].between(2015, 2024)]
price_trend = price_data.groupby('Year')['Food Price Index'].mean().reset_index()
plt.figure(figsize=(10,6))
plt.plot(price_trend['Year'], price_trend['Food Price Index'], 
         marker='o', linestyle='-', color='royalblue')
plt.xlabel("Year", fontsize=13, weight="bold")
plt.ylabel("Food Price Index", fontsize=13, weight="bold")
plt.title("Trend of Food Price Index (2015–2024)", fontsize=15, weight="bold")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(price_trend['Year'])
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

a=pd.read_csv("climate_change_food_supply_india.csv")
plt.figure(figsize=(12,9))
selected_event = ['Drought', 'Flood', 'Cyclone', 'Heatwave']
a_selected = a[a['Climate Event'].isin(selected_event)]
sns.boxplot(x=a_selected['Climate Event'], y=a_selected['Food Price Index'], color="#ffbb78")
plt.xlabel("Disaster Type", fontweight="bold", size=12)
plt.ylabel("Food Price Index", fontweight="bold", size=12)
plt.title("Effects on disaster on Food Price Index", fontweight="bold", size=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

a=pd.read_csv("climate_change_food_supply_india.csv")
selected_crops = ['Wheat', 'Rice', 'Maize', 'Cotton']
a_selected = a[a['Crop Type'].isin(selected_crops)]
plt.figure(figsize=(12,9))
sns.violinplot(x='Crop Type', y='Crop Yield (ton/ha)', data=a_selected, color="orange")
plt.xlabel("Crop Type", fontweight="bold", size=12)
plt.ylabel("Crop Yield", fontweight="bold", size=12)
plt.title("", fontweight="bold", size=15)
plt.show()
#“Among the four crops, Wheat shows the highest resilience with relatively stable and high yields, 
#while Rice and Maize exhibit more variability, and Cotton shows lower and less consistent yields.”

```
