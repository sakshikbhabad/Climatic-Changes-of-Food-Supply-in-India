```python
import pandas as pd
 #Load the dataset into a Pandas DataFrame
data=pd.read_csv('climate_change_food_supply_india_2020.csv')
#Display the first 5 records.
data
```

```python
#Find the total number of rows and columns.
a=data.shape
print("Total no.of rows and columns present in dataset", a)
#Display all column names.
col=data.columns
print("Names of the columns",col)
```

```python
#Check data types of each column.
datatype=data.dtypes
print("Type of DataTypes Present in each of column", datatype)
#Identify missing values in the dataset.
misval=data.isnull().sum()
print("No.of missing values present in dataset",misval)
```

```python
#Count the number of unique regions.
countuni=data['Region'].nunique()
print("The no.of Unique values in dataset are:",countuni)
#List all unique crop types.
unival=data['Region'].unique()
print("The Unique Values present in dataset are:",unival)
```

```python
#Filter records where Climate Event = Normal.
normal=data[data['Climate Event']=='Normal']
print("Records with Climate Event with Normal values\n",normal)
```

```python
#List crops affected by Flood.
crops_flood=data[data['Climate Event']=='Flood']['Crop Type']
print("The following are the list of crops which are affected by the Flood\n",crops_flood)
```

```python
#Find records where Rainfall > 1000 mm.
rainfall=data[data['Rainfall (mm)']>1000]
print("The following contains the records where Rainfall > 1000 mm\n",rainfall)
```

```python
#Show regions with Avg Temperature above 30°C.
avg_temp=data[data['Avg Temp (°C)']>30]['Region']
print("Regions with Avg Temperature above 30°C\n",avg_temp)
```

```python
#Filter data where Crop Yield < 3 ton/ha.
crop_yield=data[data['Crop Yield (ton/ha)']< 3]
print("Records with Crop Yied < 3 ton/ha\n", crop_yield)
```

```python
#Find all records from Uttar Pradesh.
records_up=data[data['Region']=='Uttar Pradesh']
print("Records from Uttar Pradesh\n",records_up)
```

```python
#Simple Aggregations
#Calculate the average crop yield.
average_crop_yield=data['Crop Yield (ton/ha)'].mean()
print("The average of crop yield is\n", average_crop_yield)
```

```python
#Find the maximum and minimum rainfall.
max_rainfall=data['Rainfall (mm)'].max()
min_rainfall=data['Rainfall (mm)'].min()
print("The Maximum value of Rainfall is\n", max_rainfall)
print("The Minimum value of Rainfall is\n", min_rainfall)
```

```python
#Compute average Food Price Index.
avg_food_price_index=data['Food Price Index'].mean()
print("The average value of Food Price Index is\n",avg_food_price_index)
```

```python
#Count the number of records per Climate Event.
no_climate_event=data['Climate Event'].value_counts()
print("The no.of records are present per Climate Event are",no_climate_event)
```

```python
#Find total area affected (lakh ha).
total_area_affected=data['Area Affected (lakh ha)'].sum()
print("The total area affected is",total_area_affected)
```

```python
#MODERATE LEVEL QUESTIONS
#Data Cleaning
#Convert Date column to datetime format.
data['Modified_Date']=pd.to_datetime(data['Date'])
print("Date Column to datetime format\n", data.head())
```

```python
#Standardize Severity Level values (Low, Medium, High)
def std_severity_level(severity):
    if severity in["High"]:
        return 1
    elif severity in["Medium"]:
        return 2
    elif severity in["Low"]:
        return 3

data['New_Severity_level']=data['Severity Level'].apply(std_severity_level)
data.head()
```

```python
# Round Avg Temp and Crop Yield to 2 decimal places
round_avg_temp=data['Avg Temp (°C)'].round(2)
print("The Round of values for Avg Temp column\n",round_avg_temp)

round_crop_yield=data['Crop Yield (ton/ha)'].round(2)
print("The Round of values for Crop yield Column\n",round_crop_yield)
```

```python
#Rename columns to snake_case format.
import string as str
data.columns=(data.columns.str.lower().str.replace(" ", "_"))
print(data.columns)
#Check for duplicate rows and remove them
```

```python
#Check for duplicate rows and remove them.
duplicate_rows=data.duplicated()
print("Duplicate Rows\n",duplicate_rows)
```

```python
removing_duplicates=data.drop_duplicates()
print("Removal of Duplicates\n",removing_duplicates)
```

```python
#Feature Engineering
#Create a column Yield_Category:
#Low (< 2.5 ton/ha)
#Medium (2.5–3.5 ton/ha)
#High (> 3.5 ton/ha)
def fun_yield(y_c):
    if y_c< 2.5:
        return "Low"
    elif 2.5<=y_c <=3.5:
        return "Medium"
    else:
        return "High"

data['Yield_Category']=data['crop_yield_(ton/ha)'].apply(fun_yield)
data.head()
```

```python
#Create a column Extreme_Rainfall:
#• Yes → Rainfall > 1000 mm
#• No → otherwise
def fun_rain(rain):
    if rain> 1000:
        return "Yes"
    else:
        return "No"
data['Extreme_Rainfall']=data['rainfall_(mm)'].apply(fun_rain)
data.head()
```

```python
#Create a column Price_Impact:
#• High → Food Price Index > 130
#• Low → otherwise
def fun_price(price):
    if price >130:
        return "High"
    else:
        return "Low"
data['Price_Impact']=data['food_price_index'].apply(fun_price)
data
```

```python
#Create a column Climate_Severity_Score:
#• Low = 1, Medium = 2, High = 3
def severity_level(sever):
    if sever in["Low"]:
        return 1
    elif sever in["Medium"]:
        return 2
    elif sever in["High"]:
        return 3
data['Climate_Severity_Score']=data['severity_level'].apply(severity_level)
data
```

```python
#ADVANCED–MODERATE QUESTIONS
#Grouping & Aggregation
#Find average crop yield by crop type.
average_crop_yield=data.groupby('crop_type')['crop_yield_(ton/ha)'].mean()
print("The Average Of Crop Yield by Crop Type is\n",average_crop_yield)
```

```python
#Calculate average rainfall by climate event.
average_rainfall=data.groupby('climate_event')['rainfall_(mm)'].mean()
print("The Average Rainfall by Climate Event is\n",average_rainfall)
```

```python
#Find total area affected by region.
total_area=data.groupby('region')['area_affected_(lakh_ha)'].sum()
print("The Total Area Affected By Region\n",total_area)
```

```python
#Compute mean Food Price Index by severity level.
average_food_price=data.groupby('severity_level')['food_price_index'].mean()
print("The mean of Food Price Index by Severity Level is\n",average_food_price)
```

```python
#Find average temperature by crop type.
average_temperature=data.groupby('crop_type')['avg_temp_(°c)'].mean()
print("The Average Temperature by Crop Types are\n",average_temperature)
```

```python
# Trend & Pattern Analysis
#Does higher rainfall always result in higher crop yield?
import matplotlib.pyplot as plt
x=data['rainfall_(mm)']
y=data['crop_yield_(ton/ha)']
plt.scatter(x, y, color="blue")
plt.xlabel("Rainfall(mm)")
plt.ylabel("Crop Yield(ton/ha)")
plt.show()
```

```python
#Which climate event causes the lowest average yield?
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
import seaborn as sns
x=data['climate_event'].unique()
height=data.groupby('climate_event')['crop_yield_(ton/ha)'].mean()
plt.bar(x, height, color="brown")
plt.xlabel('Climate Event', fontsize=8, fontweight="bold")
plt.xticks(rotation=90)
plt.ylabel('Average Values of Crop Yield Per Climate Event', fontsize=8, fontweight="bold")
plt.title("Comparison Between Avg.values of Crop Yield per Climate Event", fontsize=10, fontweight="bold")
plt.show()
```

```python
#Compare crop yield under Normal vs Extreme events.
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data['event_type'] = np.where(data['climate_event'] == 'Normal','Normal','Extreme')

sns.boxplot(data=data,x='event_type',y='crop_yield_(ton/ha)')

plt.title('Crop Yield under Normal vs Extreme Events')
plt.show()

```

```python
#Identify crops most affected by Floods and Droughts.
import matplotlib.pyplot as plt
data_flood_drought=data.loc[(data['climate_event']=='Flood')| (data['climate_event']=='Drought')]
a=data_flood_drought.groupby(['crop_type', 'climate_event'])['crop_yield_(ton/ha)'].mean().unstack()
a.plot(kind='bar', figsize=(10,5))
plt.xlabel('Crop Type')
plt.ylabel('Average Crop Yield (ton/ha)')
plt.title('Crops Most Affected by Floods and Droughts')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Check if High severity events increase food prices.
import seaborn as sns

sns.boxplot(data=data, x='severity_level', y='food_price_index')
plt.title('Food Price Index across Severity Levels')
plt.show()


```
